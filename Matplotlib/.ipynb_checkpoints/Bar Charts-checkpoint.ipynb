This bar graph is overlapping each other hence we can make it separate by using Numpy.
we first import numpy

```python
import numpy as np
```

We then use the arrange function **np.arange(len(x))**

```python
z=np.arange(len(ages_x))
```

```python
width=0.25
plt.bar(z-0.25,py_dev_y,width=width,color='yellow') 
plt.bar(z,js_dev_y,width=width) 
plt.bar(z+0.25,dev_y,width=width) 
plt.title('Median salary in USD by age (THIS IS TITLE)') #This is the label of Y axis


plt.xlabel('This is X axis')  #This is the label of X axis
plt.ylabel('This is Y axis')  #This is the label of Y axis



plt.show()
```
